En el punto tres (03) se solicita lo siguiente "De haber fechas, deberán tener el formato AAAA-mm-dd". Por lo que se realiza una funcion para posteriormente ser reutilizable por todos los Dataframe en los que se esta trabajando. -

```python
def date_format(dataframe: pd.DataFrame, date_column: str): # Define una función llamada date_format() que acepta dos argumentos: un objeto DataFrame y el nombre de la columna de fecha que se va a formatear.
    dataframe[date_column] = dataframe[date_column].str.strip() # Remueve los espacios en blanco al inicio y al final de las fechas en la columna especificada.
    dataframe[date_column] = pd.to_datetime(dataframe[date_column], format='%B %d, %Y') # Utilizamos la función pd.to_datetime() de Pandas para convertir las fechas en la columna especificada al formato de fecha de Pandas.
    dataframe[date_column] = dataframe[date_column].dt.strftime('%Y-%m-%d') # Utilizamos la función dt.strftime() de Pandas para formatear las fechas en la columna especificada en el formato "yyyy-mm-dd"
```

En el punto cuatro (04) se solicita que todos los campos esten en minuscula. Por lo que se realiza una funcion para posteriormente ser reutilizable por todos los Dataframe en los que se esta trabajando. -

```python
def convertir_a_minúsculas(df):
    # La función "convertir_a_minúsculas" toma un DataFrame como argumento.
    
    df = df.applymap(lambda x: x.lower() if type(x) == str else x)
    # El método "applymap" se usa para aplicar una función a cada elemento del DataFrame.
    # En este caso, la función es una lambda que convierte cada elemento a minúsculas si es una cadena,
    # y lo deja sin cambios si no lo es.
    
    return df
    # La función devuelve el DataFrame modificado.
```

En el punto cinco (05), solicitan que "El campo duration debe convertirse en dos campos: duration_int y duration_type", por lo que se crea una funcion que sea reutilizable para que cumpla con lo solicitado en los cuatros Dataframe a trabajar. -

```python
# Definir la función que recibe un DataFrame como entrada
def split_duration_column(df):
    
    # Extraer información de la columna "duration"
    duration_info = df['duration'].str.extract(r'(\d+)\s*(min|season)')
    # Utilizamos el método 'str.extract' para extraer la información de la columna 'duration'.
    # La expresión regular r'(\d+)\s*(min|season)' busca una o más ocurrencias de números (\d+), 
    # seguida de cero o más espacios en blanco (\s*), seguida de la palabra "min" o "season" (min|season).
    # Esta expresión regular identificará cualquier duración en minutos o en temporadas.
    # El resultado es un DataFrame con dos columnas: 'duration_int' (el número de minutos o temporadas) y 
    # 'duration_type' (si la duración es en minutos o en temporadas). -
    
    # Renombrar las columnas del DataFrame "duration_info"
    duration_info.columns = ['duration_int', 'duration_type']
    # Renombramos las columnas del DataFrame 'duration_info' a 'duration_int' y 'duration_type' para que el DataFrame 
    # tenga nombres de columnas más claros.-
    
    # Eliminar la columna "duration" del DataFrame original "df"
    df.drop('duration', axis=1, inplace=True)
    # Eliminamos la columna original 'duration' del DataFrame 'df' utilizando el método 'drop'. 
    # El argumento 'axis=1' especifica que queremos eliminar una columna (y no una fila).
    # Utilizamos el argumento 'inplace=True' para que los cambios se realicen directamente en el DataFrame 'df',
    # en lugar de devolver un nuevo DataFrame sin la columna 'duration'. -
    
    # Combinar el DataFrame original "df" con el DataFrame "duration_info"
    df = pd.concat([df, duration_info], axis=1)
    # Se combina el DataFrame original ("df") con el DataFrame "duration_info" utilizando el método "concat()" de pandas. 
    # El argumento "axis=1" indica que se está concatenando a lo largo de las columnas. El resultado es un nuevo DataFrame 
    # con las columnas "duration_int" y "duration_type" agregadas al final. -
    
    # La función devuelve el DataFrame resultante después de realizar todas las transformaciones descritas anteriormente. -
    return df
```
